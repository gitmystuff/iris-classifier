## 1. Data Collection & Loading

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

raw = load_iris()
df = pd.DataFrame(raw.data, columns=raw.feature_names)
df['species'] = pd.Categorical.from_codes(raw.target, raw.target_names)
df.head()
```

## 2. Exploratory Data Analysis (EDA)

```python
df.describe()
```

```python
df.groupby('species').mean(numeric_only=True)
```

```python
df.corr(numeric_only=True)
```

## 3. Data Prep: Train/Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    raw.data, raw.target, test_size=0.2, random_state=42, stratify=raw.target
)
X_train.shape, X_test.shape
```

## 4. Model Building (first pass)

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)
```

## 5. Quick Evaluation

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
```
